```python
# For Math opperations
import numpy as np

# For reading data from outside files
import pandas as pd

# For plotting data
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
%matplotlib inline
```

 ## Input Data

```python
# Allows user to choose what opperating system they are using as well as which manifold they want to look at
# This means we can establish a relative file path to the data they want to observe

System = input("Input Operating System; W for Windows, M for Mac or Linux: ")
Shape = input("Input shape; C for Circle, S for Sphere, T for torus: ")
```

```python
# Takes the user input and creates the folder address where the data is stored

if Shape == 'C' or 'c':
    if System == 'W' or 'w':
        Address = r'..\\Data\\Expected Maxima\\Circle\\'
    if System == 'M' or 'm':
        Address = r'..//Data//Expected Maxima//Circle//'
elif Shape == 'S' or 's':
    if System == 'W' or 'w':
        Address = r'..\\Data\\Expected Maxima\\Sphere\\'
    if System == 'M' or 'm':
        Address = r'..//Data//Expected Maxima//Sphere//'
elif Shape == 'T' or 't':
    if System == 'W' or 'w':
        Address = r'..\\Data\\Expected Maxima\\Torus\\'
    if System == 'M' or 'm':
        Address = r'..//Data//Expected Maxima//Torus//'
```

```python
# min N, Max N, step-size
N_Data = ['500','1950','10']

# min S, max S, step-size
S_Data = ['0.0','0.5','.001']
```

```python
# Creates a variable for the .format() function which calls specified files in the data folder

if System == 'W' or 'w':
    Files = Address + '\\{}_s0.0-0.5.csv'

if System == 'M' or 'm':
    Files = Address + '//{}_s0.0-0.5.csv'
```

```python
# Lists storing the relevant data
nvals = []
svals = []
mvals = []

for n in range(int(N_Data[0]), int(N_Data[1])+int(N_Data[2]), int(N_Data[2])):
    filenew = Files.format(n)
    
    colnames=['n', 's','m'] 
    numbers = pd.read_csv(filenew, names=colnames)
    
    for m in range(len(numbers['n'])):
        nvals.append(numbers['n'][m])
        svals.append(numbers['s'][m])
        mvals.append(numbers['m'][m])

nSorted = np.array(nvals, dtype=object)
sSorted = np.array(svals, dtype=object)
mSorted = np.array(mvals, dtype=object)
```

## Fixed S

```python
# For some reason trying to create mlist in the function above breaks things
nlist = []
slist = []
mlist = []

for n in range(int(N_Data[0]), int(N_Data[1])+int(N_Data[2]), int(N_Data[2])):
    filenew = Files.format(n)
    
    colnames=['n', 's','m'] 
    numbers = pd.read_csv(filenew, names=colnames)
    
    nlist.append(list(numbers['n']))
    slist.append(list(numbers['s']))
    mlist.append(list(numbers['m']))
```

```python
S = float(input("Input Desired s Value: "))
```

```python
# The index function tells us the index of our S value so we can ensure a proper zip with the future m values
Index = slist[0].index(S)
```

```python
nFixed = []
mFixed = []

for s in range(len(slist)):
    if S in slist[s]:
        nFixed.append(nlist[s][Index])
        mFixed.append(mlist[s][Index])
        
nFixed = np.array(nFixed)
mFixed = np.array(mFixed)
```

```python
def fixs(x,a,b):
    return a*(np.log(x)**b)

popt,tcov = curve_fit(fixs, nFixed, mFixed)
```

```python
%matplotlib inline
f = plt.figure(figsize = (12,8))
ax1 = plt.subplot(111)
ax1.set_ylim(0,int(np.ceil(max(mFixed)))+1)


ax1.scatter(nFixed, mFixed, c='blue')
ax1.plot(nFixed, fixs(nFixed, *popt), c='red')

print("For a*(ln(n)^b) you have a,b values: ")
print(popt)
plt.show()
```

# 3D Plot

```python
%matplotlib notebook
f = plt.figure(figsize = (12,12))
ax1 = plt.axes(projection = '3d')

ax1.set_xlabel('n')
ax1.set_ylabel('s')
ax1.set_zlabel('E[Mn]')
ax1.scatter3D(nSorted, sSorted, mSorted, cmap='coolwarm', c=mSorted)

plt.show()
```
