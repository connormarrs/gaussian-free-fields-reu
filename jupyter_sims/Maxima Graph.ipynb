```python
%matplotlib notebook
import matplotlib.pyplot as plt
import json
```

```python
# This program is designed to create a 3d graph relating s and n to E[M_n].

def create_sbyn_graph(file):
    # The json file this accepts must be of the type where s is not fixed.
    f = open(file)
    data = json.load(f)
    nvals = [] # To be graphed on the x-axis.
    svals = [] # To be graphed on the y-axis
    mvals = [] # To be graphed on the z-axis.
    for n in data['info']['n'].keys():
        for s in data['info']['n'][n]['s'].keys():
            mean = data['info']['n'][n]['s'][s]["meanOfMaxima"]
            # By appending these together, we ensure proper indexing for zipping.
            nvals.append(int(n))
            svals.append(float(s))
            mvals.append(mean)
    ax = plt.axes(projection = '3d')
    ax.set_xlabel('n')
    ax.set_ylabel('s')
    ax.set_zlabel('E[Mn]')
    ax.scatter3D(nvals, svals, mvals, cmap='coolwarm')
```

```python
create_sbyn_graph('d2_s0.0-0.5_n1-50.json')
```
