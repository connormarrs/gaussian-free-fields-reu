```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
colnames=['n', 'E(n)','E'] 
exp = pd.read_csv('exp_sample.csv', names=colnames)

print(exp)
```

```python
f = plt.figure(figsize = (16,8))
ax = f.add_subplot(121)
ax1 = f.add_subplot(122)

p = ax.scatter(exp["n"],exp["E(n)"])
p = ax1.hist2d(exp["n"],exp["E(n)"],bins=50)
p = sns.jointplot(exp["n"],exp["E(n)"])
p.axisbelow: True
```

```python
f = plt.figure(figsize = (16,8))
ax = f.add_subplot(121)
ax1 = f.add_subplot(122)

p = ax.scatter(exp["n"],exp["E"])
p = ax1.hist2d(exp["n"],exp["E"],bins=30)
p = sns.jointplot(exp["n"],exp["E"])
p.axisbelow: True
```

```python
f = plt.figure(figsize = (10,10))
ax = f.add_subplot(111)

sns.histplot(exp["E(n)"], kde=True, bins = 50)

plt.show()
```
