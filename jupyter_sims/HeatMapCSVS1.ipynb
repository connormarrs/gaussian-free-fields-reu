```python
# These Functions construct the addresses of all the files in the folder

def filenumber(n):
    number = address + '\\' + n
    return number

def filename(n ,s):
    name = filenumber(n) + '_' + s + '.csv'
    return name
```

```python
# This creates a list of all data addresses the user specified
Files = []

for n in range(len(N)):
    for s in range(len(S)):
        Files.append(filename((N[n]),S[s]))
```

### Get CSV Data

```python
# Takes a csv file and reads all the data
def getdata(file):
    colnames = ['k','data']
    File_Data = pd.read_csv(file, names=colnames)
    return File_Data
```

```python
# creates a list of all the data
All_Data = []

for x in range(len(Files)):
    All_Data.append(getdata(Files[x]))
```

```python
# creates a list with only the trial data
Trial_Data  = []

for x in range(len(All_Data)):
    Trial_Data.append(list(All_Data[x]['data']))
    
for y in range(len(Trial_Data)):
    Trial_Data[y].append(0)
```

### Getting Ready to Graph

```python
# Creates a list of theta values and a list of zero values for graphing

def getTheta(n):
    theta = [(2*np.pi*k)/n for k in range(0, n+1)]
    return theta

def getLine(n):
    line = [0 for k in range(n+1)]
    return line
```
