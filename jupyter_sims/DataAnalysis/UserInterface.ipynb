```python
def floatValidate(potentialFloat):
    if potentialFloat != '':
        try: 
            float(potentialFloat)
        except: 
            return False
    return True


def intValidate(potentialInt):
    if potentialInt != '':
        try: 
            int(potentialInt)
        except:
            return False
    return True
```

```python
def heatMaps(*args):
    nList = []
    numberList = []
    for digit in nVals.get().strip().replace(' ', ''):
        if digit != " " and digit != ',':
            numberList.append(digit)
        elif digit == ",":
            n = "".join(numberList)
            nList.append(n)
            numberList = []
    n = "".join(numberList)
    nList.append(n)
    sList = []
    floatList = []
    for digit in sVals.get().strip().replace(' ', ''):
        if digit != " " and digit != ',':
            floatList.append(digit)
        elif digit == ",":
            s = "".join(floatList)
            sList.append(s)
            floatList = []
    s = "".join(floatList)
    sList.append(s)
    sListPadded = []
    for s in sList:
        index = s.find(".")
        if index != -1:
            sPadded = s.ljust(index+7, "0")
        else:
            sPadded = s.ljust(6, "0")
        sListPadded.append(sPadded)    
    if surface.get() == "Circle (S1)":
        if OS.get() == "Windows":
            files = heatMaps_S1.getFilesWindows(address, nList, sListPadded)
        else:
            files = heatMaps_S1.getFiles(address, nList, sListPadded)
        trialData, badFiles = heatMaps_S1.getTrialData(files)
        heatMaps_S1.graphHeatMaps(trialData, sListPadded, nList, badFiles)
    if surface.get() == "Torus (T2)":
        if OS.get() == "Windows":
            files = heatMaps_T2.getFilesWindows(address, nList, sListPadded)
        else:
            files = heatMaps_T2.getFiles(address, nList, sListPadded)
        trialData, badFiles = heatMaps_T2.getTrialData(files)
        heatMaps_T2.graphHeatMaps(trialData, sListPadded, nList, badFiles)
    if surface.get() == "Sphere (S2)":
        if OS.get() == "Windows":
            files = heatMaps_S2.getFilesWindows(address, nList, sListPadded)
        else:
            files = heatMaps_S2.getFiles(address, nList, sListPadded)
        allThetas, allPhis, trialData, badFiles = heatMaps_S2.getTrialData(files)
        heatMaps_S2.graphHeatMaps(allThetas, allPhis, trialData, sListPadded, nList, badFiles)
```

```python
def sValidate(sVals):
    periods = 0
    length = 0
    for s in sVals.strip().replace(' ', ''):
        try:
            float(s)
            if periods == 1:
                length += 1
            if length > 6:
                return False
        except:
            if s == ",":
                periods = 0
                length = 0
            elif s == ".":
                periods += 1
                if periods > 1:
                    return False
            else:
                return False
    return True


def getSVals(*args):
    global sValsPrompt
    global sValsEntry
    try:
        sValsPrompt.destroy()
        sValsEntry.destroy()
    except: pass
    global sVals
    sVals = StringVar()
    sValsEntry = Entry(master, validate = 'key', validatecommand = (master.register(sValidate), '%P'),
                       textvariable = sVals, font=('calibre', 10, 'normal'))
    sValsPrompt = Label(master, text = "Enter your desired s values, separated by commmas:")
    sValsPrompt.pack()
    sValsEntry.pack()
    sValsEntry.bind('<Return>', heatMaps)
```
